```python
# Question 3A

# create a function Missing_Cols
def Missing_Cols(dataframe):

# sum the null values in each column and create a series
    nullsum = pd.Series(dataframe.isnull().sum())
# convert series to dataframe
    nulldf = pd.DataFrame(nullsum)
# rename column of sum of null values to Missing
    nulldf = nulldf.rename(columns = {0:"Missing"})
# return the desired output
    return nulldf
Missing_Cols(NHL_Goalies)
```

```python
# Question 3B

# create a function Missing_Rows
def Missing_Rows(dataframe):
# sum the null values in each row and append a new column called Missing Values
    dataframe['Missing_Values'] = dataframe.isnull().sum(axis=1)
# return the dataframe
    return dataframe
Missing_Rows(NHL_Goalies)
```

```python
# Question 4

GAA_Goalies = np.asarray(NHL_Goalies["GAA"])
GAA_workhorse = np.asarray(workhorse["GAA"])


fig, ax = plt.subplots(1,2)
st = fig.suptitle("Goals Against Average", fontsize="x-large")
ax[0].hist(GAA_Goalies, alpha = 0.5, color = 'red')
ax[1].hist(GAA_workhorse, alpha = 0.5, color = 'navy')
ax[0].set_title("NHL Goalies GAA")
ax[1].set_title("NHL Workhorse Goalies GAA")

plt.show()
```

```python
# Question 5

# subset the data for injuries not known
subset = NHL_Goalies[NHL_Goalies["Injuries"].isnull()]

# merge data using left and right indexes as join keys
mergedDF=pd.merge(subset, GAA_, left_index = True, right_index = True, how='inner')

# keep desired columns
mergedDF=mergedDF.loc[:, ['Last Name_x', 'First Name_x','Team_x','Cntry', 'SV%_x', 'GA_x', 'GAA_x','TOI']]

# groupby country and aggregate means, mins, max of kept columns
mergedDF.groupby('Cntry').agg(['mean', 'min', 'max'])

# create dataframe of TOI and write it to a csv file
TOI_dataframe = mergedDF.groupby('Cntry').agg({'TOI':['mean','min','max']})
TOI_dataframe.to_csv('TOI_2017.csv')
```

```python
# Question 6

# function that takes dataframe and column as an input
def interquartile(dataframe, column):
    
# acquire all column names
    colnames = dataframe.columns
# acquire all numeric columns
    newdf = dataframe._get_numeric_data()
    
# if the column is not valid print error
    if column not in colnames:
        print("You have not entered a valid column name")
# if the column is not numeric print error
    elif column not in newdf: 
        print("You have not entered a numeric column name")
# if valid and numeric continue to find iqr
    else:
        quartone = dataframe[column].quantile(0.25)
        quartthree = dataframe[column].quantile(0.75)
        IQR = (dataframe[column] <= quartone) | (dataframe[column] >= quartthree)
        return dataframe[IQR]

# this will produce the correct code
interquartile(NHL_Goalies, "Wt")

# this will produce a valid column name error
# interquartile(NHL_Goalies, "Kiran")

# this will produce a numeric column name error
# interquartile(NHL_Goalies, "Cntry")
```
